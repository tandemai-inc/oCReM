# oCReM Tutorial

This tutorial demonstrates the basic operations of oCReM (optimized Chemical Reassembly), including:
- Fragmentation of input molecular structures
- Saving fragmentation results to database

## Molecular Fragmentation <a name='fragmentation'></a>

oCReM provides fragmentation capabilities that break down molecules into structural fragments. Let's explore the fragmentation process with some example molecules.

First, let's create a file for test.

```python
with open('test.smi', 'w') as f:
    f.write('CC(=O)Oc1ccccc1C(=O)O\n')
```

Fragmentation of input structures:

```python
import os

path_to_fragmentation = os.path.join(os.path.dirname(os.getcwd()), "ta_gen/bin/fragmentation.py")

os.system(
    f"export PYTHONPATH={os.path.dirname(os.getcwd())}:$PYTHONPATH && "
    f"python {path_to_fragmentation} --input test.smi --out test_frag.csv --mode 0 --ncpu 10 --radius 3"
)
```

```python
import os

path_to_fragmentation = os.path.join(os.path.dirname(os.getcwd()), "ta_gen/bin/fragmentation.py")

os.system(
    f"export PYTHONPATH={os.path.dirname(os.getcwd())}:$PYTHONPATH && "
    f"python {path_to_fragmentation} --input test.smi --out test_frag.txt --mode 0 --ncpu 10 --radius 3"
)
```

Available parameters for fragmentation:
- `-i, --input_file`: Input compound file
- `--sep`: Delimiter for input file (default: ',')
- `-c, --ncpu`: Number of CPUs to use
- `--chunk_size`: Chunk size for processing input file (default: 100)
- `-m, --mode`: Fragmentation mode (0=heavy atoms only, 1=heavy atoms and hydrogens,  2=hydrogen atoms only)
- `--max_heavy_atoms`: Maximum number of heavy atoms in cores
- `--radius`: Radius of molecular context
- `--use_db`: Flag to save results to database
- `--db_type`: Database type (sqlite or postgres)
- `--db_path`: Database path
- `--ini_file`: PostgreSQL configuration file
- `--reset_db`: Reset database before insert data.
- `--debug`: Debug mode. Save fragmented molecules. It will be true, if not use_db.
- `-o, --out`: Output file for fragmented molecules.
- `-sep_out`: Delimiter for output file (default: ',')

Fragmentation of input structures and save results to sqlite database:

```python
import os

path_to_fragmentation = os.path.join(os.path.dirname(os.getcwd()), "ta_gen/bin/fragmentation.py")

os.system(
    f"export PYTHONPATH={os.path.dirname(os.getcwd())}:$PYTHONPATH && "
    f"python {path_to_fragmentation} --input test.smi --out test_frag.txt --mode 0 --ncpu 10 --radius 3 --use_db --db_type sqlite --db_path test.db"
)
```

Fragmentation of input structures and save results to PostgreSQL database:

```python
import os

path_to_fragmentation = os.path.join(os.path.dirname(os.getcwd()), "ta_gen/bin/fragmentation.py")

os.system(
    f"export PYTHONPATH={os.path.dirname(os.getcwd())}:$PYTHONPATH && "
    f"python {path_to_fragmentation} --input test.smi --out test_frag.txt --mode 0 --ncpu 10 --radius 3 --use_db --db_type postgres --ini_file test.ini"
)
```
